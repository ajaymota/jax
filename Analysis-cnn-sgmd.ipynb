```python
import pandas as pd
from matplotlib import pyplot as plt
import jax.numpy as np
import seaborn as sns
```

```python
df_cnn_sgmd0 = pd.read_csv("results/cnn_sgmd_K0.csv")
df_cnn_sgmd1 = pd.read_csv("results/cnn_sgmd_K1.csv")
df_cnn_sgmd2 = pd.read_csv("results/cnn_sgmd_K2.csv")
df_cnn_sgmd3 = pd.read_csv("results/cnn_sgmd_K3.csv")
df_cnn_sgmd4 = pd.read_csv("results/cnn_sgmd_K4.csv")
df_cnn_sgmd5 = pd.read_csv("results/cnn_sgmd_K5.csv")
df_cnn_sgmd6 = pd.read_csv("results/cnn_sgmd_K6.csv")
df_cnn_sgmd7 = pd.read_csv("results/cnn_sgmd_K7.csv")
df_cnn_sgmd8 = pd.read_csv("results/cnn_sgmd_K8.csv")
df_cnn_sgmd9 = pd.read_csv("results/cnn_sgmd_K9.csv")
```

```python
p = sns.lineplot(data=df_cnn_sgmd0, dashes=False)
p.set_title("Sigmoid Kernel 1")
p.set(xlabel='epochs', ylabel='eigenvalues')
p.legend(bbox_to_anchor=(1.05, 1), title="n-th eigenvalue", loc=2, borderaxespad=0.)

plt.show()
```
