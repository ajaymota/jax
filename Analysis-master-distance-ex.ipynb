```python
import pandas as pd
import numpy as np
```

```python
def calc_dist(df1, df2):
    x = df1.corrwith(df2)
    x[0] = np.linalg.norm(df1['0'].to_numpy()-df2['0'].to_numpy())
    x[1] = np.linalg.norm(df1['1'].to_numpy()-df2['1'].to_numpy())
    x[2] = np.linalg.norm(df1['2'].to_numpy()-df2['2'].to_numpy())
    x[3] = np.linalg.norm(df1['3'].to_numpy()-df2['3'].to_numpy())
    x[4] = np.linalg.norm(df1['4'].to_numpy()-df2['4'].to_numpy())
    return x
```
