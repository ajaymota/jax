```python
def calc_distance(my_str_df):
    
    my_df = globals()[my_str_df]
    s1 = my_df['Sigmoid']
    s2 = my_df['Tanh']
    s3 = my_df['RelU']
    
    my_df_04 = globals()[my_str_df + "_04"]
    s1_04 = my_df_04['Sigmoid']
    s2_04 = my_df_04['Tanh']
    s3_04 = my_df_04['RelU']
    
    my_df_06 = globals()[my_str_df + "_06"]
    s1_06 = my_df_06['Sigmoid']
    s2_06 = my_df_06['Tanh']
    s3_06 = my_df_06['RelU']
    
    my_df_08 = globals()[my_str_df + "_08"]
    s1_08 = my_df_08['Sigmoid']
    s2_08 = my_df_08['Tanh']
    s3_08 = my_df_08['RelU']
    
    max_dist = max(
        max((s1-s2).abs()),
        max((s2-s3).abs()),
        max((s3-s1).abs()),
        
        max((s1_04-s2_04).abs()),
        max((s2_04-s3_04).abs()),
        max((s3_04-s1_04).abs()),
        
        max((s1_06-s2_06).abs()),
        max((s2_06-s3_06).abs()),
        max((s3_06-s1_06).abs()),
        
        max((s1_08-s2_08).abs()),
        max((s2_08-s3_08).abs()),
        max((s3_08-s1_08).abs())
    )
    
    max_norm = np.linalg.norm([max_dist] * len(s1))
    
    n1 = np.linalg.norm(s1 - s2) / max_norm
    n2 = np.linalg.norm(s2 - s3) / max_norm
    n3 = np.linalg.norm(s3 - s1) / max_norm
    
    n1_04 = np.linalg.norm(s1_04 - s2_04) / max_norm
    n2_04 = np.linalg.norm(s2_04 - s3_04) / max_norm
    n3_04 = np.linalg.norm(s3_04 - s1_04) / max_norm
    
    n1_06 = np.linalg.norm(s1_06 - s2_06) / max_norm
    n2_06 = np.linalg.norm(s2_06 - s3_06) / max_norm
    n3_06 = np.linalg.norm(s3_06 - s1_06) / max_norm
    
    n1_08 = np.linalg.norm(s1_08 - s2_08) / max_norm
    n2_08 = np.linalg.norm(s2_08 - s3_08) / max_norm
    n3_08 = np.linalg.norm(s3_08 - s1_08) / max_norm
    
    return  (n1, n2, n3), \
            (n1_04, n2_04, n3_04), \
            (n1_06, n2_06, n3_06), \
            (n1_08, n2_08, n3_08)
```

```python
# returns err value between the final epochs
def get_err(fcn1, fcn2):
    e1 = pd.read_csv("results/" + fcn1 + "/cnn_test_class_loss.csv")
    e2 = pd.read_csv("results/" + fcn2 + "/cnn_test_class_loss.csv")
    
    err = abs(e1.iloc[8]['0'] - e2.iloc[8]['0'])
    
    return err

def get_err_tuple():
    
    dataset = 'mnist'
    combs = [('sgmd', 'tanh'), ('tanh', 'relu'), ('relu', 'sgmd'),
             ('dropout/0.4/sgmd', 'dropout/0.4/tanh'), 
             ('dropout/0.4/tanh', 'dropout/0.4/relu'), 
             ('dropout/0.4/relu', 'dropout/0.4/sgmd'),
             ('dropout/0.6/sgmd', 'dropout/0.6/tanh'),
             ('dropout/0.6/tanh', 'dropout/0.6/relu'),
             ('dropout/0.6/relu', 'dropout/0.6/sgmd'),
             ('dropout/0.8/sgmd', 'dropout/0.8/tanh'), 
             ('dropout/0.8/tanh', 'dropout/0.8/relu'), 
             ('dropout/0.8/relu', 'dropout/0.8/sgmd')]

    err_list = []

    for comb in combs:
        err_list.append(get_err(comb[0], comb[1]))
    
    return  (err_list[0], err_list[1], err_list[2]), \
            (err_list[3], err_list[4], err_list[5]), \
            (err_list[6], err_list[7], err_list[8]), \
            (err_list[8], err_list[10], err_list[11])
```
