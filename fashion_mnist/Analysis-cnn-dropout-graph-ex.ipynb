# Distances between Eigenvalues

```python
import copy

def calc_dist(df1, df2):
    x = df1.corrwith(df2)
    x[0] = np.linalg.norm(df1['0'][-10:].to_numpy()-df2['0'][-10:].to_numpy())
    x[1] = np.linalg.norm(df1['1'][-10:].to_numpy()-df2['1'][-10:].to_numpy())
    x[2] = np.linalg.norm(df1['2'][-10:].to_numpy()-df2['2'][-10:].to_numpy())
    x[3] = np.linalg.norm(df1['3'][-10:].to_numpy()-df2['3'][-10:].to_numpy())
    x[4] = np.linalg.norm(df1['4'][-10:].to_numpy()-df2['4'][-10:].to_numpy())
    y = copy.deepcopy(x)
    y[0] = np.linalg.norm(np.asarray(x))
    y[1] = np.linalg.norm(np.asarray(x))
    y[2] = np.linalg.norm(np.asarray(x))
    y[3] = np.linalg.norm(np.asarray(x))
    y[4] = np.linalg.norm(np.asarray(x))
    return y
```
