```python
import pandas as pd
from matplotlib import pyplot as plt
import jax.numpy as np
import seaborn as sns
```

```python
df_relu = pd.read_csv("results/relu_L2.csv")
df_sgmd = pd.read_csv("results/sgmd_L2.csv")
df_relu_t = pd.read_csv("results/relu_train_loss.csv")
df_sgmd_t = pd.read_csv("results/sgmd_train_loss.csv")
```

```python
df_relu.head()
```

```python
df_sgmd.head()
```

```python
p = sns.lineplot(data=df_relu, dashes=False)
p.set_title("ReLU Output Layer")
p.set(xlabel='epchocs', ylabel='eigenvalues')
p.legend(bbox_to_anchor=(1.05, 1), title="n-th eigenvalue", loc=2, borderaxespad=0.)

plt.show()
```

```python
p = sns.lineplot(data=df_sgmd, dashes=False)
p.set_title("Sigmoid Output Layer")
p.set(xlabel='epchocs', ylabel='eigenvalues')
plt.legend(bbox_to_anchor=(1.05, 1), title="n-th eigenvalue", loc=2, borderaxespad=0.)
plt.show()
```

```python
df_sgmd_t['Sigmoid'] = df_sgmd_t
df_sgmd_t['ReLU'] = df_relu_t
df_sgmd_t.drop('0', axis=1, inplace=True)
```

```python
p = sns.lineplot(data=df_sgmd_t, dashes=False)
p.set_title("Error Sigmoid v/s ReLU")
p.set(xlabel='epchocs', ylabel='ms-error')
plt.legend(bbox_to_anchor=(1.05, 1), title="functions", loc=2, borderaxespad=0.)
plt.show()
```
