```python
import pandas as pd
from matplotlib import pyplot as plt
import jax.numpy as np
import seaborn as sns
```

```python
activator = "tanh"
dropout_rate = 0.8
```

```python
df_cnn_relu_train = pd.read_csv("results/dropout/" + str(dropout_rate) + "/" + activator + "/cnn_train_loss.csv")
df_cnn_relu_test = pd.read_csv("results/dropout/" + str(dropout_rate) + "/" + activator + "/cnn_test_loss.csv")
df_cnn_relu_test_class = pd.read_csv("results/dropout/" + str(dropout_rate) + "/" + activator + "/cnn_test_class_loss.csv")
```

```python
df_cnn_relu_train['train'] = df_cnn_relu_train
df_cnn_relu_train['test'] = df_cnn_relu_test
df_cnn_relu_train['classification'] = df_cnn_relu_test_class
df_cnn_relu_train.drop('0', axis=1, inplace=True)
```

```python
p = sns.lineplot(data=df_cnn_relu_train, dashes=False)
p.set_title("Error - train v/s test v/s classification")
p.set(xlabel='epchocs', ylabel='error')
plt.legend(bbox_to_anchor=(1.05, 1), title="type", loc=2, borderaxespad=0.)
plt.show()
```

```python
df_cnn_relu0 = pd.read_csv("results/dropout/" + str(dropout_rate) + "/" + activator + "/cnn_K0.csv")
df_cnn_relu1 = pd.read_csv("results/dropout/" + str(dropout_rate) + "/" + activator + "/cnn_K1.csv")
df_cnn_relu2 = pd.read_csv("results/dropout/" + str(dropout_rate) + "/" + activator + "/cnn_K2.csv")
df_cnn_relu3 = pd.read_csv("results/dropout/" + str(dropout_rate) + "/" + activator + "/cnn_K3.csv")
df_cnn_relu4 = pd.read_csv("results/dropout/" + str(dropout_rate) + "/" + activator + "/cnn_K4.csv")
df_cnn_relu5 = pd.read_csv("results/dropout/" + str(dropout_rate) + "/" + activator + "/cnn_K5.csv")
df_cnn_relu6 = pd.read_csv("results/dropout/" + str(dropout_rate) + "/" + activator + "/cnn_K6.csv")
df_cnn_relu7 = pd.read_csv("results/dropout/" + str(dropout_rate) + "/" + activator + "/cnn_K7.csv")
df_cnn_relu8 = pd.read_csv("results/dropout/" + str(dropout_rate) + "/" + activator + "/cnn_K8.csv")
df_cnn_relu9 = pd.read_csv("results/dropout/" + str(dropout_rate) + "/" + activator + "/cnn_K9.csv")
```
